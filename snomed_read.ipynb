```python
def parse_file(filename, first_row_header=True, columns=None):
    with open(filename, encoding='utf-8') as f:
        entities = [[n.strip() for n in line.split('\t')] for line in f]
        return pd.DataFrame(entities[1:], columns=entities[0] if first_row_header else columns)
```

## SNOMED CT Design
### SNOMED CT Components
SNOMED CT is a clinical terminology containing concepts with unique meanings and formal logic based definitions organised into hierarchies.

SNOMED CT content is represented into 3 main types of components:

Concepts representing clinical meanings that are organised into hierarchies.

Descriptions which link appropriate human readable terms to concepts

Relationships which link each concept to other related concepts

```python

int_terms = parse_file(f'{int_terminology}/sct2_Concept_Snapshot_INT_20240101.txt')
be_terms = parse_file(f'{be_ext_terminology}/sct2_Concept_Snapshot_BE1000172_20231115.txt')
terms = pd.concat([int_terms, be_terms])
active_terms = terms[terms.active == '1'] # active concepts are represented with 1
```

```python
int_desc = parse_file(f'{int_terminology}/sct2_Description_Snapshot-en_INT_20240101.txt')
be_desc = parse_file(f'{be_ext_terminology}/sct2_Description_Snapshot-fr_BE1000172_20231115.txt')
descs = pd.concat([int_desc, be_desc])
active_descs = descs[descs.active == '1']
```

```python
int_relat = parse_file(f'{int_terminology}/sct2_Relationship_Snapshot_INT_20240101.txt')
be_relat = parse_file(f'{be_ext_terminology}/sct2_Relationship_Snapshot_BE1000172_20231115.txt')
relat = pd.concat([int_relat, be_relat])
active_relat = relat[relat.active == '1']
```
